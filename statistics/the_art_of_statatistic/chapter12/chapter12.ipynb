## Ілюзія "Золотого стандарту"   
Розділ починається з абсурдного прикладу: у 2011 році авторитетний журнал   
опублікував дослідження, яке нібито доводило існування екстрасенсорних здібностей    
(ESP) у студентів. Автор отримав $p$-value $= 0.01$ і заявив про відкриття.    
Це яскрава ілюстрація проблеми: наука та бізнес настільки сліпо моляться на $P < 0.05$,   
що готові публікувати магію, якщо вона пройшла математичний фільтр.   
Це призвело до Кризи відтворюваності (Reproducibility crisis).

## Парадокс Гельмана (Бінарна пастка)  

Коли масштабний проєкт спробував повторити 100 класичних досліджень, лише 36% показали      повторний "успіх".   
Преса одразу заявила, що 64% науки — це брехня. Але статистик    
Ендрю Гельман пояснює глибшу проблему: різниця між "значущим" і "незначущим"    
сама по собі не є значущою.  
Як це розуміти: Якщо Оригінальний тест показав приріст $+5\%$ ($P = 0.04$ — "Успіх!"),   
а Повторний тест показав $+4\%$ ($P = 0.06$ — "Провал!"), вони насправді не суперечать   
один одному. Обидва показують позитивний тренд. Але через штучну лінію у $0.05$   
люди роблять хибний висновок, що другий тест повністю спростував перший.

## Де все ламається: Помилки в циклі PPDAC

Навіть без злого наміру, статистика розвалюється на базових етапах дизайну 
та виконання.     
Приклади з практики:  
- **Plan (Планування): Зручна вибірка та потужність.** Проводити опитування лише серед тих,   
до кого легко додзвонитися (Selection bias). Або запускати A/B тест на занадто малій аудиторії   (Low Power), через що неможливо відрізнити реальний ефект від шуму.  
- **Data (Дані): Ігнорування прихованих факторів.** Оцінювати ефективність гомеопатії,     ігноруючи ефект плацебо (відсутність сліпого тестування), або не враховувати вік та стать   
при аналізі хвороб (Confounders).
- **Analysis (Аналіз): Технічний та методологічний брак.** Іноді це не складна математика, а просто загублена колонка в Excel (як у статті Рейнхарта-Рогоффа) або пропущений мінус у коді моделі ризиків (що коштувало фонду AXA Rosenberg $242 млн штрафів). А іноді це хибний метод — наприклад, тестувати групи окремо замість прямого порівняння їх між собою (Test of interaction).
- **Conclusion (Висновки): P-Hacking.** Зібрати 20 різних метрик, почекати поки одна з них випадково покаже "ріст", і гордо відзвітувати лише про неї, приховавши 19 фейлів (Selective reporting).

### Multiple Testing (множинне тестування) та P-hacking  

Коли ми фіксуємо рівень значущості $\alpha = 0.05$, ми погоджуємося на $5\%$ ймовірність   зробити хибнопозитивну помилку (Type I error) в одному тесті.  
Але якщо ми тестуємо відразу 20 різних метрик (вага, холестерин, тиск, якість сну тощо),   
як це робив Боханнон, ймовірність знайти хоча б один "статистично значущий" ефект   
випадково обчислюється так:

$$P(\text{хоча б один хибний плюс}) = 1 - (1 - \alpha)^m$$Де $m$ — кількість тестів. Для 20 тестів:$$P = 1 - (1 - 0.05)^{20} = 1 - 0.95^{20} \approx 1 - 0.358 = 0.642$$  
Тобто, навіть якщо шоколад не робить нічого, у нас є $64\%$ шансів отримати "успішний"  
A/B тест

```python
import numpy as np
import scipy.stats as stats


np.random.seed(13)
n_metrics = 20
sample_size = 15

false_positives = 0
for i in range(n_metrics):
    group_a = np.random.normal(loc=0, scale=1, size=sample_size)
    group_b = np.random.normal(loc=0, scale=1, size=sample_size)

    t_stat, p_value= stats.ttest_ind(group_a, group_b)

    if p_value < 0.05:
        false_positives += 1
        print(f"Metric {i+1}: False positive detected (p-value: {p_value:.4f})")

print(f"Total false positives: {false_positives} out of {n_metrics}")
```

статистичні заяви, використовуючи сумнівні методи інтерпретації та комунікації.
