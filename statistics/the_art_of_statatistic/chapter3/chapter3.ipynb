## The Path of Inductive Inference

The route that data must take so that we can trust it.  

1. **Sample** - the specific people we interviewed  
2. **Study Population** - the group from which we could actually  
   take people, for example people from a telephone database  
3. **Target Population** - those we actually want to know something about,  
   for example all British adults

**What is the problem?**  
Biases arise at every step  
For example, studies on sexual partners:  
Selection bias: Perhaps only more relaxed people agree to answer  
intimate questions?
Non-response bias:  
Those who refused may be radically different from those who answered.

## The Bell-Shaped Curve /  Distribution

**What is a Normal Distribution?**

It is a perfectly symmetrical "bell" shape. Imagine that we  
measured the weight of 10,000 babies.
Most of the values ​​will cluster around the mean.  
The further we move away from the center   
(towards very small or very large weights), the fewer such babies we will find.

**Main characteristics:**  
Mean ($\mu$): The peak of the bell. This is where the largest number    
of values ​​are found.  
Standard Deviation ($\sigma$): This is the ruler we use to measure the "spread".   
It tells us how wide the bell is.  

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set ggplot style for an R-like appearance
plt.style.use('ggplot')

# Define parameters for the Standard Normal Distribution
mu = 0
sigma = 1

# Generate x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Create the figure
fig, ax = plt.subplots(figsize=(12, 7))

# Plot the main curve
ax.plot(x, y, color='black', linewidth=2)


# 1 Standard Deviation (68%)
x1 = np.linspace(mu - sigma, mu + sigma, 100)
ax.fill_between(x1, norm.pdf(x1, mu, sigma), color='#348ABD', alpha=0.8)

# 2 Standard Deviations (95%)
x2_left = np.linspace(mu - 2*sigma, mu - sigma, 100)
x2_right = np.linspace(mu + sigma, mu + 2*sigma, 100)
ax.fill_between(x2_left, norm.pdf(x2_left, mu, sigma), color='#E24A33', alpha=0.8)
ax.fill_between(x2_right, norm.pdf(x2_right, mu, sigma), color='#E24A33', alpha=0.8)

# 3 Standard Deviations (99.7%)
x3_left = np.linspace(mu - 3*sigma, mu - 2*sigma, 100)
x3_right = np.linspace(mu + 2*sigma, mu + 3*sigma, 100)
ax.fill_between(x3_left, norm.pdf(x3_left, mu, sigma), color='#988ED5', alpha=0.8)
ax.fill_between(x3_right, norm.pdf(x3_right, mu, sigma), color='#988ED5', alpha=0.8)

# ADD TEXT INSIDE THE CURVE
ax.text(-0.5, 0.15, '34.1%', ha='center', color='white', fontweight='bold', fontsize=12)
ax.text(0.5, 0.15, '34.1%', ha='center', color='white', fontweight='bold', fontsize=12)

ax.text(-1.5, 0.05, '13.6%', ha='center', color='white', fontweight='bold', fontsize=11)
ax.text(1.5, 0.05, '13.6%', ha='center', color='white', fontweight='bold', fontsize=11)

ax.text(-2.5, 0.015, '2.1%', ha='center', color='white', fontweight='bold', fontsize=10)
ax.text(2.5, 0.015, '2.1%', ha='center', color='white', fontweight='bold', fontsize=10)


# 68% Arrow
ax.annotate('', xy=(-1, 0.25), xytext=(1, 0.25), arrowprops=dict(arrowstyle="<->", color='black', lw=1.5))
ax.text(0, 0.26, '68%', ha='center', va='bottom', fontsize=12, fontweight='bold')

# 95% Arrow
ax.annotate('', xy=(-2, 0.08), xytext=(2, 0.08), arrowprops=dict(arrowstyle="<->", color='black', lw=1.5))
ax.text(0, 0.09, '95%', ha='center', va='bottom', fontsize=12, fontweight='bold')

# 99.7% Arrow
ax.annotate('', xy=(-3, 0.02), xytext=(3, 0.02), arrowprops=dict(arrowstyle="<->", color='black', lw=1.5))
ax.text(0, 0.03, '99.7%', ha='center', va='bottom', fontsize=12, fontweight='bold')


ax.set_title('Normal Distribution: The 68-95-99.7 Rule', fontsize=16, fontweight='bold', pad=20)
ax.set_ylabel('Probability Density', fontsize=12)

# Customize X-axis to show standard deviations (mu, sigma)
ax.set_xticks([-3, -2, -1, 0, 1, 2, 3])
ax.set_xticklabels([r'$-3\sigma$', r'$-2\sigma$', r'$-1\sigma$', r'$\mu$', r'$+1\sigma$', r'$+2\sigma$', r'$+3\sigma$'], fontsize=14)

# Remove the default grid for a cleaner look
ax.grid(False)

plt.tight_layout()
plt.show()
```

**The 68–95–99.7 Rule**

These are the magic numbers of statistics. In any normal distribution:  

68% of all values ​​lie within 1 standard deviation of the mean.  
95% of values ​​lie within 2 standard deviations. (This is considered the "normal range").  
99.7% of values ​​lie within 3 standard deviations. Anything outside these limits  
is a great rarity (Outliers).

**The example of a 6-pound baby is a classic.**
  

The mother worries that the weight is "below average."   
Spiegelhalter uses this to explain:
Average: It's just a point. Almost no one weighs exactly that much.  
Normal Distribution: Most values ​​cluster around the center, and  
the further away from it, the less likely they are to occur.  
Percentiles: If a child is in the 15th percentile, that means 15% of children  
weigh less and 85% weigh more. This is "below average," but still within the normal range.

## Literal vs Metaphorical

**Literal Population:** These are real people (e.g., all residents of Ukraine).  
We take a sample of them.  

**Metaphorical Population:**  
What if we had data on all cases (e.g., all hospital deaths in a year)?  
Spiegelhalter says we can think of them as a “sample” of all possible ways  
that year could have gone. This helps us understand whether what happened  
was a pattern or just a coincidence.

## **Summary**

**Random Sample** - This is the "gold standard". If we select people randomly,  
we minimize the risk that our sample will be biased.  
**Parameters vs Statistics:**  
**Statistics:** These are the numbers we get from our sample   
(e.g., the mean of 100 people surveyed).  
**Parameters:** These are the "true" numbers for the entire population    
that we are trying to guess.  
All the data: Even if we know everything about the past, we use statistics  
to build a model for the future or to understand a hidden process.
