## **Correlation Does Not Imply Causation**

**The most famous example of the section:**   
Education and the laws of the brain.  
A study showed that people with higher education are more likely  
to have brain diseases.  

Do books cause cancer? Of course not.  
What is the secret? This is a confounding factor.  
People with higher education are quite wealthy, often go to hospitals and  
have access to high-quality diagnostics (MRI). Therefore, they simply find  
marriages more often than those who do not consult doctors.  

## **What is "Causation" really?**

For statisticians, causality is the answer to the question:  
"What would happen if we intervened?"  
If we made a person take statins, would their risk of having  
a heart attack decrease? If so, that's causality.  

## **Gold standard: Randomized controlled trials (RCTs)**

To prove causality, we use Randomized Controlled Trials (RCT).  
    
**Randomization:** We divide people into two groups completely randomly  
(like flipping a coin). This ensures that both groups are the same in all respects  
(age, gender, wealth).

**Blinding:** The patient (and preferably the doctor) does not know whether they are taking  
the real drug or the "placebo".  

**Intention-to-treat:** The analysis is carried out on the original groups, even if  
someone stopped taking the drug. This helps to maintain the "purity" of randomization.  

The example with statins proved their effectiveness precisely through RCTs.  
But the study of prayer (Is prayer effective?) showed that when people know that they  
are being prayed for, it sometimes even worsens the outcome due to the stress of waiting
for recovery.

## **Observational data and the "Waitrose Effect"**

When we can’t do an experiment (for example, we can’t make people smoke for 20 years  
for an experiment), we use Observational data – just observing the world.  

Example from Waitrose: Having a high-end supermarket nearby adds £36,000 to the value  
of a house.  

Causation or Association? It’s just Association. The supermarket doesn’t make an area rich;  
on the contrary, Waitrose is just very clever in choosing areas where people are already rich  
to open a store there.

## **Why do old men have big ears? (Lurking factors)**

This is a great example of how we can get our conclusions wrong:  
It's not because ears grow throughout life.  

It's a Lurking factor: ears get longer due to gravity and loss of skin elasticity  
(the aging process), not due to the mere fact of "being old."

## **Summary**

**Intervention:** This is the main test of causality. If changing A changes B,  
there is a relationship.  
**Confounders:** These are "third extras" that create the illusion of a relationship  
where there is none.  
**Adjusting:** Statisticians have mathematical methods to "remove" the effects of  
age or wealth from the analysis to see the net effect.  
**Judgment:** Statistics alone never say 100% "this is the cause." They give numbers,  
and the final decision is up to the person and their common sense.
