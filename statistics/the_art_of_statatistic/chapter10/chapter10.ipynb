## Boys vs. Girls: The First Test of the Hypothesis in History
The author begins the chapter with the story of John Arbuthnot, Queen Anne’s physician. He collected data on birth rates in London   over 82 years (from 1629 to 1710) and calculated the ratio of boys born per 100 girls. He discovered a striking pattern: **slightly more boys were always born.** This ratio ranged from 101 to 116, with an average of 105 boys for every 100 girls. Arbuthnot wanted to prove that this was not a coincidence but a fundamental law. His logic was ingenious in its simplicity: If we assume that the chance of a boy or a girl being born is exactly 50:50 (like flipping a coin), what is the probability that the “coin” will land in favor of boys for 82 years in a row? Mathematically, this is $(1/2)^{82}$. This is such a minuscule number (24 zeros after the decimal point!) that it simply cannot be a statistical coincidence. He concluded that nature (or, as he wrote, “the wise Creator”) deliberately produces more males to compensate for their higher mortality rate over the course of a lifetime. **In essence, he conducted the first-ever test of a statistical hypothesis.**
