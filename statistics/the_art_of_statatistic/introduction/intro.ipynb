## Why we need Statictics ?

The author does not start with complex abstract mathematical formulas,  
but with a real-life example of how a killer doctor killed his patients, 
and how statistics are used to represent data (his victims died in the afternoon, 
most of them were elderly women, which is why they discovered these anomalies in the data).

```python
import pandas as pd
import matplotlib.pyplot as plt

# set style ggplot
plt.style.use('ggplot')

# read the data
df = pd.read_csv('shipman-confirmed-victims.csv')

counts = df.groupby(['ageBracket','gender2']).size().unstack(fill_value=0)

# plot the data
ax = counts.plot(kind='bar', stacked=True, figsize=(12, 7), 
                 color=['#1F77B4', '#FF7F0E'], width=0.8)



for c in ax.containers:
    labels = [int(v.get_height()) if v.get_height() > 0 else "" for v in c]
    ax.bar_label(c, labels=labels, label_type='center', color='white', fontweight='bold')


plt.title('Amount of victims by age group and gender', fontsize=14, pad=15)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')

df = pd.read_csv('shipman-times-x.csv')

df.set_index('Hour', inplace=True)

# Line chart
plt.figure(figsize=(12, 6))

plt.plot(df.index, df['Shipman'], label='Shipman', color='#d62728', linewidth=2, marker='o')
plt.plot(df.index, df['Comparison'], label='Comparison', color='#1f77b4', linewidth=2, marker='o')

plt.title('Deaths by Hour of Day', fontsize=14)
plt.xlabel('Hour', fontsize=12)
plt.ylabel('% Deaths', fontsize=12)

plt.xticks(range(0, 24)) 
plt.legend()
plt.tight_layout()

plt.show()
```

## PPDAC Cycle

The author shows us that statistics is not a set of
mathematical tricks, but a process of solving problems.

Problem: Clearly formulate the question
Plan: Determine what data we need
Data: Collect, clean the data
Analysis: Find patterns, visualize, build models
Conclusion: Interpret the results
