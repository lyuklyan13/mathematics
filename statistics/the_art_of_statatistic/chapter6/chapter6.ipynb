## **Finding Patterns**

Where before we were looking for the mean, now we are looking for rules.  
Using the example of the passengers of the Titanic, the author shows  
how an algorithm can study data to predict: "Who will survive?"  
**Features:** Gender, age, cabin class.  
**Outcome:** Alive or dead.

## **Classification Trees**

This is one of the simplest and most understandable algorithms.  
It works as a series of "yes/no" questions.  

**Question 1:** Is this a man? (If yes, go left, if no, go right).  
**Question 2:** How old is he?  
**Question 3:** What grade is he in?  

Spiegelhalter says that such trees are very convenient because  
we can see the logic of decision-making (unlike complex artificial intelligence).

## **Over-fitting — The main trap**

This is a very important concept.

**Over-fitting:** This is when an algorithm becomes too "smart" and  
remembers every random detail in your data (noise) instead of finding  
the general rule (signal).  
**Consequence:** Such an algorithm works perfectly on old data,  
but completely fails on new data.  
**Solution:** Always split the data into a Training set (what we learn from)  
and a Test set (what we test against).

## **How to evaluate the quality of an algorithm?**

The author gives the example of a weather forecast. If a forecaster says  
"70% chance of rain," how do you know if he's right?  
**Discrimination:** Can the algorithm distinguish one group from another?  
**Calibration:** If we predicted a 70% chance of rain on 100 different days,  
then it should actually rain on 70 of those days.  
**Brier Score:** A special formula for measuring the accuracy of probabilistic forecasts.

## **Algorithms in Medicine: Predict Program**

Spiegelhalter mentions the Predict tool, which helps women with breast cancer  
decide whether they need chemotherapy.  
The algorithm takes thousands of case histories and gives a prediction:  
"What are your chances of living another 10 years with and without therapy?"  
This is an example of how statistics become personalized medicine.  

## **Summary**

**Noise vs Signal:**  The main challenge is not to adjust to random fluctuations  
in the data (Noise).  
**Transparency:** The more complex an algorithm (like neural networks) is,  
the harder it is to understand why it made the decision it did.  
Sometimes it is better to choose a simpler model that we understand.  
**Artificial Intelligence (AI):** It is not magic. It is just very complex algorithms  
built on huge amounts of data that have the same limitations as regular statistics.
