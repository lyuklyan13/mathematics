```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

np.random.seed(14)
n_points = 1000

correlations = [0.0, 0.10, 0.50, 0.90]

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
axes = axes.flatten()

x = np.random.normal(0, 1, n_points)
z = np.random.normal(0, 1, n_points)

for i, rho in enumerate(correlations):
    y = rho * x + np.sqrt(1 - rho**2) * z
    
    axes[i].scatter(x, y, alpha=0.5, color='#348ABD', edgecolor='white', s=40)
    axes[i].set_title(f'Correlation: {int(rho * 100)}%', fontsize=16, fontweight='bold')
    
    axes[i].set_xlim(-4, 4)
    axes[i].set_ylim(-4, 4)
    axes[i].set_aspect('equal', adjustable='box')
    
    axes[i].set_xlabel('Variable X')
    axes[i].set_ylabel('Variable Y')

plt.tight_layout()
plt.show()
```
