```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr


# 1. Generate non-linear monotonic data ( exponential growth)
x = np.linspace(1, 10, 100)
y = np.exp(x)  # Y grows much faster than X

# 2. Calculate both correlation coefficients
# pearsonr and spearmanr return a tuple: (correlation_coefficient, p-value)
pearson_corr, _ = pearsonr(x, y)
spearman_corr, _ = spearmanr(x, y)

# 3. Plot the data to visualize the difference
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', alpha=0.7)

# Add title with the calculated correlation values
plt.title(
    f'Non-linear Monotonic Relationship\n'
    f'Pearson (Linear): {pearson_corr:.2f}\n'
    f'Spearman (Rank): {spearman_corr:.2f}', 
    fontsize=14, 
    fontweight='bold'
)

plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y (Exponential)', fontsize=12)

plt.tight_layout()
plt.show()
```
