```python
import math

def manual_correlation(x, y):
    # Check if lists have the same length
    if len(x) != len(y) or len(x) == 0:
        raise ValueError("Lists must be of the same length and not empty.")
    
    n = len(x)
    
    # 1. Calculate the mean (expected value) for X and Y
    mean_x = sum(x) / n
    mean_y = sum(y) / n
    
    # Variables to store the sums for covariance and variances
    covariance = 0.0
    variance_x = 0.0
    variance_y = 0.0
    
    # 2. Iterate through the data to calculate the sums
    for xi, yi in zip(x, y):
        # Difference from the mean
        dx = xi - mean_x
        dy = yi - mean_y
        
        # Add to covariance
        covariance += dx * dy
        
        # Add to variances
        variance_x += dx ** 2
        variance_y += dy ** 2
        
    # 3. Handle division by zero (if all values in a list are identical)
    if variance_x == 0 or variance_y == 0:
        return 0.0
        
    # 4. Calculate final correlation
    correlation = covariance / math.sqrt(variance_x * variance_y)
    
    return correlation
```

```python
# TEST THE FUNCTION

x_data = [1, 2, 3, 4, 5]
y_data = [2, 4, 5, 4, 5]

result = manual_correlation(x_data, y_data)
print(f"Correlation: {result:.4f}")
```
