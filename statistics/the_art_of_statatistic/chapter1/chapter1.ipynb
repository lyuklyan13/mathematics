## Bristol Heart Surgery

The author gives the example of pediatric heart surgery.  
The problem was that the mortality rate in Bristol seemed higher than in other cities.  

The gist: Here we encounter Binary variables for the first time - a yes/no question  
(did the patient survive or not).

Lesson: Counts alone don't tell us much. We need to look at Proportions.  
If 10 children died in one hospital and 5 in another, it doesn't mean the first hospital is worse,   
until we know how many operations they performed in total.

## Categorical Variables

This is data that is divided into groups.  
For example:  
Survivor / Died (Binary).  
Blood type (A, B, AB, O).  

Type of surgery.  
Spiegelhalter emphasizes that visualization of such  
data (charts, tables) should be very careful.

## Relative vs. Absolute Risk

This is one of the most famous parts of the book.   
It explains how the media creates panic.  

Relative Risk: The newspapers write: "Eating bacon increases the risk of cancer by 18%!"  
That sounds scary.  

Absolute Risk: In fact, the risk of bowel cancer is about 5% (5 in 100 people).  
If you eat bacon every day, it will increase to 6% (6 in 100 people).  

The math here is simple: An increase from 5% to 6% is an increase of 18%  
from the original value. But "1% more" doesn't sound as hype as "18% more".

## Expected Frequencies

Spiegelhalter suggests that the best way to communicate  
is to use "living people" instead of percentages.  

Instead of saying "6% risk," it's better to say, "Out of 100 people who don't eat bacon,  
5 will get sick; out of 100 people who eat it every day, 6 will get sick."  
This gives a real sense of scale.

## Summary

Framing: You can say "90% survival" (positive) or "10% mortality" (negative).  
    The emotional impact is different, even though the number is the same.  

Odds Ratios: This is a tricky thing for scientific articles.  
The author advises against using them for the general public, because people always  
confuse them with probability. Odds of 1 to 1 (1:1) is a 50% probability. It's confusing.  

How does 85% to 87% become a "20% increase"? This is a manipulation of context.  
For example, if we are not talking about the total number, but about the "excess"  
part above a certain threshold. Always ask: "20% of WHAT?"

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')

df_pie = pd.read_excel('child-heart-pie-chart-x.xlsx')

df_pie.rename(columns={df_pie.columns[0]: 'Hospital'}, inplace=True)

plt.figure(figsize=(10, 8))

plt.pie(df_pie['% ops carried out in each hospital'], 
        labels=df_pie['Hospital'], 
        autopct='%1.1f%%', 
        startangle=140, 
        wedgeprops={'edgecolor': 'white', 'linewidth': 1}) 

plt.title('Share of surgeries in each hospital', fontsize=14, pad=20)
plt.tight_layout()

plt.show()
```

```python
df_bar = pd.read_csv('child-heart-survival-x.csv')

df_bar_sorted = df_bar.sort_values('Operations', ascending=True)

plt.figure(figsize=(12, 7))

bars = plt.barh(df_bar_sorted['Hospital'], df_bar_sorted['Operations'], color='#1f77b4')

plt.title('Total Operations by Hospital', fontsize=14)
plt.xlabel('Number of Operations', fontsize=12)
plt.ylabel('Hospital', fontsize=12)

for bar in bars:
    plt.text(bar.get_width() + 10,  
             bar.get_y() + bar.get_height()/2, 
             f'{int(bar.get_width())}', 
             va='center', fontsize=10)

plt.tight_layout()

plt.show()
```
