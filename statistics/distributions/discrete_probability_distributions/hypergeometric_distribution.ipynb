## **Theoretical Foundation:**
Unlike the Binomial distribution, the Hypergeometric distribution models the probability  
of $k$ successes in $n$ draws from a finite population of size $N$ without replacement,  
where the population contains exactly $K$ successes.

$$P(X = k) = \frac{\binom{K}{k} \binom{N-K}{n-k}}{\binom{N}{n}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# N: total population, K: total successes in population, n: number of draws
N_tot, K_succ, n_draws = 100, 20, 10
dist_hyper = hypergeom(M=N_tot, n=K_succ, N=n_draws)

print(f'Mean:\t {dist_hyper.mean()} | Variance:\t {dist_hyper.var():.3f}')

x_vals = np.arange(0, n_draws + 1)
fig, (ax1, ax2 )= plt.subplots(1, 2, figsize=(12, 5))
ax1.vlines(x_vals, 0, dist_hyper.pmf(x_vals), colors='c', lw=4)
ax1.set_title('Hypergeometric PMF')
ax2.step(x_vals, dist_hyper.cdf(x_vals), where='post', color='c', lw=2)
ax2.set_title('Hypergeometric CDF')
plt.show()
```

## **Practical Application:**
It is essential in ecological capture-recapture studies to estimate wildlife population sizes,  
and in legal auditing to determine the probability of finding fraudulent documents in a small sample  
drawn from a finite corpus without replacement.

## **Interesting Fact:**
As the total population size $N \to \infty$, the dependency created by sampling  
without replacement becomes negligible. Consequently, the Hypergeometric distribution  
seamlessly converges into the Binomial distribution.
