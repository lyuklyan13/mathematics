## **Theoretical Foundation:**
The Geometric distribution models the number of Bernoulli trials required to achieve  
the first success, where each trial has a success probability $p$. The probability mass   
decays exponentially with each subsequent failure.

$$P(X = k) = (1-p)^{k-1} p$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

p_success = 0.3
dist_geom = geom(p=p_success)

print(f'Mean:\t {dist_geom.mean():.3f} | Variance:\t {dist_geom.var():.3f}')

x_vals = np.arange(1, 15)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.vlines(x_vals, 0, dist_geom.pmf(x_vals), colors='m', lw=4)
ax1.set_title(f'Geometric PMF (p={p_success})')
ax2.step(x_vals, dist_geom.cdf(x_vals), where='post', color='m', lw=2)
ax2.set_title(f'Geometric CDF (p={p_success})')
plt.show()
```

## **Practical Application:**
In reliability engineering and operations research, it models the lifespan of highly robust,  
non-degrading components (like specific solid-state electronics), where the probability of  
failure on any given day is constant

## **Interesting Fact:**
It is the only discrete probability distribution that exhibits the "memoryless" property.  
The probability of requiring additional trials does not depend on how many trials have already failed,  
echoing the continuous Exponential distribution.
