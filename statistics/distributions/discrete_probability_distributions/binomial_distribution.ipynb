## **Theoretical Foundation:**
The Binomial distribution computes the probability of obtaining exactly $k$ successes in $n$  
independent Bernoulli trials. Each trial must have a constant probability of success $p$.  
The formulation relies on the binomial coefficient to account for the combinatorics of success sequences:
$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

The expected value expands linearly as $np$, while the variance is $np(1-p)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n_trials, p_success = 20, 0.4
dist_binom = binom(n=n_trials, p=p_success)

print(f'Mean:\t {dist_binom.mean()} | Variance:\t {dist_binom.var()}')

sample_binom = dist_binom.rvs(size=1000, random_state=13)
x_vals = np.arange(0, n_trials + 1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# PMF plot
ax1.vlines(x_vals, 0, dist_binom.pmf(x_vals), colors='b', lw=4, alpha=0.7, label='PMF')
ax1.hist(sample_binom, bins=np.arange(-0.5, n_trials+1.5, 1), density=True, alpha=0.3)
ax1.set_title('Binomial PMF')

# CDF plot
ax2.step(x_vals, dist_binom.cdf(x_vals), where='post', color='b', lw=2)
ax2.set_title('Binomial CDF')

plt.show()
```

## **Practical Application:**
It is extensively deployed in quality control and conversion rate optimization.  
If an assembly line produces defective parts with a probability of 0.02,  
the binomial distribution dictates the probability that a shipped batch of 500 parts  
contains more than the acceptable threshold of defects.

## **Interesting Fact:**
According to the De Moivre-Laplace theorem, as $n$ approaches infinity,  
the discrete binomial distribution converges smoothly into a continuous normal distribution.  
This discovery historically catalyzed the formulation of the Central Limit Theorem.
