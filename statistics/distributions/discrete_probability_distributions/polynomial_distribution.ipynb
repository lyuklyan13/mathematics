## **Theoretical Foundation:**
The Multinomial distribution is the multi-dimensional generalization of the Binomial distribution.  
It models the outcome of $n$ independent trials, but each trial can result in exactly one of $k$  
fixed categories, governed by a probability vector $\mathbf{p} = (p_1, \dots, p_k)$ that sums to 1.

$$P(X_1=x_1, \dots, X_k=x_k) = \frac{n!}{x_1! \cdots x_k!} p_1^{x_1} \cdots p_k^{x_k}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multinomial

n_trials = 20
p_vec = [0.2, 0.5, 0.3]
dist_multi = multinomial(n=n_trials, p=p_vec)

# generate 5 samples
n_samples = 5
sample_multi = dist_multi.rvs(size=n_samples, random_state=42)

print("Multinomial Samples (Rows sum to 20):")
print(sample_multi)

# Visualization: Stacked bar chart
fig, ax = plt.subplots(figsize=(8, 6))

sample_indices = np.arange(1, n_samples + 1)
# variable to track the bottom position for stacking
bottom_counts = np.zeros(n_samples) 

colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
labels = ['Category 1 (p=0.2)', 'Category 2 (p=0.5)', 'Category 3 (p=0.3)']

# Build bar segments for each of the 3 categories
for i in range(len(p_vec)):
    category_counts = sample_multi[:, i]
    ax.bar(sample_indices, category_counts, bottom=bottom_counts, color=colors[i], edgecolor='black', label=labels[i])
    bottom_counts += category_counts 

# Chart formatting
ax.set_xlabel('Sample Number')
ax.set_ylabel('Counts')
ax.set_title(f'Multinomial Distribution Samples (n={n_trials})')
ax.set_xticks(sample_indices)
ax.axhline(n_trials, color='red', linestyle='--', alpha=0.7, label='Total sum (n=20)')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

## **Practical Application:**
It forms the mathematical bedrock of Natural Language Processing (NLP), specifically in  
"Bag of Words" models and Naive Bayes classifiers, where text is treated as a   
multinomial distribution over a vast categorical vocabulary.

## **Interesting Fact:**
The Multinomial distribution acts as the data-generating counterpart to the continuous  
Dirichlet distribution. In Bayesian statistics, the Dirichlet distribution is the conjugate  
prior for the Multinomial, enabling highly efficient algorithmic learning.
