## **Theoretical Foundation:**
The Zipf (or Zeta) distribution is a discrete power-law distribution.  
The PMF is inversely proportional to the rank $k$ of an element, governed by a decay parameter $s > 1$.

$$P(X=k) = \frac{1}{k^s \zeta(s)}$$

where $\zeta(s)$ is the Riemann zeta function serving as the normalization constant.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zipf

s_param = 2.0
dist_zipf = zipf(a=s_param)

x_vals = np.arange(1, 20)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.bar(x_vals, dist_zipf.pmf(x_vals), color='orange', alpha=0.7)
ax1.set_title(f'Zipf PMF (s={s_param})')
ax2.step(x_vals, dist_zipf.cdf(x_vals), where='post', color='orange')
ax2.set_title(f'Zipf CDF (s={s_param})')
plt.show()
```

## **Practical Application:**
Zipf's law is empirically ubiquitous in Data Science. It perfectly models the frequency of  
word occurrences in natural human languages, the population of cities, and the distribution  
of internet traffic to web servers, dictating caching optimization strategies.

## **Interesting Fact:**
Because it relies on the Riemann zeta function, if the shape parameter $s \le 1$, the infinite  
sum diverges. This implies that for heavily extreme cases, the distribution cannot theoretically  
exist over an infinite domain without truncation.
