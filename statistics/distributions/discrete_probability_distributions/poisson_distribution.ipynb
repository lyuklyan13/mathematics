## **Theoretical Foundation:**
The Poisson distribution expresses the probability of a given number of events occurring  
in a fixed interval of time or spatial domain, assuming these events occur with a known  
constant mean rate $\lambda$ and independently of the time since the last event.  

$$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

rate_lambda = 5.0
dist_poiss = poisson(mu=rate_lambda)

print(f'Mean:\t {dist_poiss.mean()} | Variance:\t {dist_poiss.var()}')

x_vals = np.arange(0, 15)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.vlines(x_vals, 0, dist_poiss.pmf(x_vals), colors='g', lw=4, label='PMF')
ax2.step(x_vals, dist_poiss.cdf(x_vals), where='post', color='g', lw=2)

plt.show()
```

## **Practical Application:**
It is the absolute core of queueing theory. Network engineers use it to model the number of data  
packets arriving at a router per millisecond, and epidemiologists use it to model the frequency of  
rare disease mutations in a fixed population.

## **Interesting Fact:**
A mathematically unique property of the Poisson distribution is that its expected value (mean) is  
strictly equal to its variance ($E[X] = \text{Var}(X) = \lambda$). Any empirical data where variance vastly   
exceeds the mean indicates "overdispersion," necessitating a switch to the Negative Binomial distribution.
