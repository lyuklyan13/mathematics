## **Theoretical Foundation:**

The discrete uniform distribution represents a scenario of maximum entropy for a finite state space.  
If a random variable $X$ can assume any integer value between boundaries $a$ and $b$  (inclusive)  
with absolute equiprobability, the Probability Mass Function is defined as $P(X = k) = \frac{1}{b - a + 1}$.  
The theoretical mean evaluates to $\frac{a+b}{2}$, and the variance is governed by $\frac{(b-a+1)^2 - 1}{12}$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define parameters: lower bound (inclusive) and upper bound (exclusive in scipy)
a, b = 1, 7 # Simulates a standard 6-sided die
dist_dunif = randint(low=a, high=b)

print('Numerical Characteristics of Discrete Uniform Distribution:')
print('-' * 60)
print(f'Expected Value (Mean):\t\t {dist_dunif.mean()}')
print(f'Variance:\t\t\t {dist_dunif.var()}')
print(f'Standard Deviation:\t\t {dist_dunif.std():.3f}')
print(f'Skewness:\t\t\t {dist_dunif.stats(moments="s"):.3f}')
print(f'Kurtosis:\t\t\t {dist_dunif.stats(moments="k"):.3f}')
print(f'Lower Quartile (25%):\t\t {dist_dunif.ppf(0.25)}')
print(f'Median (50%):\t\t\t {dist_dunif.median()}')
print(f'Upper Quartile (75%):\t\t {dist_dunif.ppf(0.75)}')

# Generate sample
sample_dunif = dist_dunif.rvs(size=1000, random_state=13)

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
x_vals = np.arange(a, b)

# PMF Plot
ax1.plot(x_vals, dist_dunif.pmf(x_vals), 'ro', ms=8, label='PMF')
ax1.vlines(x_vals, 0, dist_dunif.pmf(x_vals), colors='r', lw=5, alpha=0.5)
ax1.hist(sample_dunif, density=True, bins=np.arange(a, b+1)-0.5, alpha=0.4, color='blue', edgecolor='black', label='Sample Histogram')
ax1.set_title('Probability Mass Function (PMF)')
ax1.set_xlabel('k')
ax1.set_ylabel('P(X = k)')
ax1.legend()

# CDF Plot 
ax2.step(x_vals, dist_dunif.cdf(x_vals), where='post', color='r', lw=2, label='CDF')
ax2.hist(sample_dunif, density=True, cumulative=True, bins=np.arange(a, b+1)-0.5, alpha=0.4, color='blue', edgecolor='black')
ax2.set_title('Cumulative Distribution Function (CDF)')
ax2.set_xlabel('k')
ax2.set_ylabel('F(k)')
ax2.legend()

plt.suptitle('Discrete Uniform Distribution U{1, 6}', fontsize=14)
plt.tight_layout()
plt.show()
```

### **Practical Application:**


In computer science, algorithms requiring unbiased state initialization, random hashing,  
or A/B test cohort assignments heavily rely on discrete uniform distributions. It guarantees  
that no partition of the data receives an inherent mathematical bias during initialization.

## **Interesting Fact:**
Despite its simplicity, proving the true uniformity of pseudo-random number generators (PRNGs)  
on digital computers is notoriously difficult, requiring rigorous cryptographic standards  
(like the Mersenne Twister) to avoid hidden periodicities.
