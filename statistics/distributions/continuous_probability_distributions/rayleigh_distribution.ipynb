## **Theoretical Foundation:**
The Rayleigh distribution models the magnitude of a two-dimensional vector whose  
orthogonal components are independently, identically, and normally distributed   
with zero mean and equal variance.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh

dist_rayleigh = rayleigh(scale=1.0)

x_vals = np.linspace(0, 5, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_rayleigh.pdf(x_vals), color='c', lw=3)
ax1.set_title('Rayleigh Distribution PDF')
ax2.plot(x_vals, dist_rayleigh.cdf(x_vals), color='c', lw=3)
ax2.set_title('Rayleigh Distribution CDF')
plt.show()
```

## **Practical Application:**
In meteorology, it is the global standard for modeling the magnitude of wind  
speed vectors. In medical imaging, it describes the magnitude of background noise  
in Magnetic Resonance Imaging (MRI) scans, allowing machines to separate tissue  
signals from static.

## **Interesting Fact:**
If you fire bullets at a target, and the $X$ and $Y$ horizontal/vertical aiming  
errors are random normally distributed without a directional bias, the resulting  
radial Euclidean distance of the bullet holes from the exact bullseye follows  
a perfect Rayleigh distribution.
