## **Theoretical Foundation**:
The continuous uniform distribution allocates a constant probability   
density to all real numbers within a bounded interval $[a, b]$.

$$f(x) = \frac{1}{b-a} \quad \text{for } x \in [a, b]$$

The mean is $\frac{a+b}{2}$ and the variance is $\frac{(b-a)^2}{12}$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# In scipy, loc is the lower bound, scale is the width (b - a)
loc_u, scale_u = -1, 3 
dist_unif = uniform(loc=loc_u, scale=scale_u)

print(f'Mean:\t {dist_unif.mean()} | Variance:\t {dist_unif.var()}')

x_vals = np.linspace(-2, 3, 1000)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_unif.pdf(x_vals), 'r-', lw=3)
ax1.fill_between(x_vals, dist_unif.pdf(x_vals), alpha=0.2, color='red')
ax1.set_title('Continuous Uniform PDF')

ax2.plot(x_vals, dist_unif.cdf(x_vals), 'r-', lw=3)
ax2.set_title('Continuous Uniform CDF')
plt.show()
```

## **Practical Application:**
It acts as the algorithmic engine for Inverse Transform Sampling. Data scientists use  
standard continuous uniform variables $U(0,1)$ and map them through the inverse CDF of  
complex distributions to generate random variables for Monte Carlo simulations.

## **Interesting Fact:**
According to the Principle of Maximum Entropy, if the only known constraint about  
a continuous random variable is that it exists within a finite interval $[a, b]$,  
the uniform distribution is the most unbiased and information-theoretically sound choice.
