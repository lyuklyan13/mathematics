## **Theoretical Foundation:**
While not a traditional probability density function, the Dirac Delta $\delta(x - \mu)$   
characterizes a Degenerate Distribution where absolutely all probability mass is concentrated   
at a single deterministic point $\mu$. The variance is exactly zero.

```python
# Simulated as the limit of a Gaussian approaching zero variance
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu_dirac = 5.0
epsilon = 1e-4 
dist_dirac = norm(loc=mu_dirac, scale=epsilon)

x_vals = np.linspace(4.99, 5.01, 1000)
plt.figure(figsize=(6, 5))
plt.plot(x_vals, dist_dirac.pdf(x_vals), color='k', lw=2)
plt.title('Approximation of Dirac Delta Function')
plt.show()
```

## **Practical Application:**
In Bayesian statistics, it represents a prior distribution with absolute, dogmatic certainty.  
In signal processing, it is mathematically essential for sampling continuous signals into  
discrete domains (the impulse response of a system).

## **Interesting Fact:**
The Dirac Delta function is technically a "distribution" in the sense of functional analysis  
(a generalized function), not strictly a probability distribution. It operates as the limiting  
case of almost any standard continuous distribution (Normal, Cauchy, Laplace) as the scale  
parameter violently shrinks to zero.
