## **Theoretical Foundation:**
A continuous distribution characterized by the fact that the logarithm of its probability  
density function maps out a hyperbola. It yields semi-heavy tails that decay slower than   
a Normal distribution but significantly faster than a Cauchy distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import genhyperbolic

p_gh, a_gh, b_gh = 1, 2, 0
dist_gh = genhyperbolic(p=p_gh, a=a_gh, b=b_gh)

x_vals = np.linspace(-5, 5, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_gh.pdf(x_vals), color='midnightblue', lw=3)
ax1.set_title('Generalized Hyperbolic PDF')
ax2.plot(x_vals, dist_gh.cdf(x_vals), color='midnightblue', lw=3)
ax2.set_title('Generalized Hyperbolic CDF')
plt.show()
```

## **Practical Application:**
Standard Gaussian models dangerously underestimate the frequency of extreme stock market fluctuations.  
Because financial returns notoriously exhibit "fat tails", the hyperbolic distribution is widely  
employed in quantitative finance, econometrics, and Value at Risk (VaR) modeling to provide a mathematically safe buffer.  

## **Interesting Fact:**
The visual structure of the log-probability perfectly forms a hyperbola.   
Because of this, the tails of the distribution exhibit exponential decay on a logarithmic scale,  
perfectly bridging the gap between absolute stability (Normal) and catastrophic variance (Cauchy).
