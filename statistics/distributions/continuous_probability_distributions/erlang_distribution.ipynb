## **Theoretical Foundation:**
The Erlang distribution is a specialized, rigid sub-case of the Gamma distribution  
where the shape parameter $k$ is constrained to be a strictly positive integer.

$$f(x) = \frac{\lambda^k x^{k-1} e^{-\lambda x}}{(k-1)!}$$

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import erlang

k_erlang = 5 # Must be an integer
dist_erlang = erlang(a=k_erlang, scale=1.0)

x_vals = np.linspace(0, 15, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_erlang.pdf(x_vals), color='navy', lw=3)
ax1.set_title(f'Erlang PDF (k={k_erlang})')
ax2.plot(x_vals, dist_erlang.cdf(x_vals), color='navy', lw=3)
ax2.set_title(f'Erlang CDF (k={k_erlang})')
plt.show()

## **Practical Application:**
Invented by A.K. Erlang to examine the number of telephone calls concurrently  
handled by human operators, it remains fundamental in teletraffic engineering  
and server load balancing to calculate waiting times in multi-node networks.

## **Interesting Fact:**
Because $k$ is strictly an integer, the normalization constant uses the simple  
factorial $(k-1)!$ rather than the complex analytical Gamma function $\Gamma(k)$   
required by the broader Gamma distribution, accelerating computational rendering   
in massive simulations.
