## **Theoretical Foundation:**
Often translated dynamically from Cyrillic prompts as "Xi", the standard statistical nomenclature  
is the Chi ($\chi$) distribution. It characterizes the positive square root of a Chi-squared  
distributed variable. Mathematically, it is the distribution of the Euclidean distance from  
the origin in a $k$-dimensional space, where each coordinate is an independent standard normal variable.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi

df_chi = 3
dist_chi = chi(df=df_chi)

x_vals = np.linspace(0, 5, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_chi.pdf(x_vals), color='dodgerblue', lw=3)
ax1.set_title(f'Chi PDF (df={df_chi})')
ax2.plot(x_vals, dist_chi.cdf(x_vals), color='dodgerblue', lw=3)
ax2.set_title(f'Chi CDF (df={df_chi})')
plt.show()
```

## **Practical Application:**
In physics and physical chemistry, the Chi distribution with exactly 3 degrees  
of freedom strictly governs the Maxwell-Boltzmann distribution, representing  
the speed of molecules in an ideal gas.

## **Interesting Fact:**
While the Chi-squared distribution is famous for testing variances,  
the Chi distribution geometrically represents actual physical distances (the $L^2$ norm).
