## **Theoretical Foundation:**
The Inverse Gaussian distribution (frequently called the Wald distribution) is driven  
by a mean $\mu$ and shape parameter $\lambda$. It models continuous stochastic processes   
approaching a boundary.

$$f(x) = \sqrt{\frac{\lambda}{2\pi x^3}} \exp\left( -\frac{\lambda(x-\mu)^2}{2\mu^2 x} \right)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import invgauss

mu_invgauss = 1.0
dist_invgauss = invgauss(mu=mu_invgauss)

x_vals = np.linspace(0.01, 4, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_invgauss.pdf(x_vals), color='seagreen', lw=3)
ax1.set_title('Wald (Inverse Gaussian) PDF')
ax2.plot(x_vals, dist_invgauss.cdf(x_vals), color='seagreen', lw=3)
ax1.set_title('Wald (Inverse Gaussian) CDF')

plt.show()
```

## **Practical Application:**
The Wald distribution dictates the first passage time in a Brownian motion process.  
In industrial manufacturing, it acts as the foundation for Sequential Probability Ratio Tests (SPRT),  
where production lines are continuously monitored to sequentially detect shifts in quality   
control without fixed sample sizes.

## **Interesting Fact:**
The "Inverse" in the name does not imply that $1/X$ is Gaussian. Rather, it reflects   
an inversion of logic: while a Gaussian distribution describes the distance traveled  
in a fixed amount of time by Brownian motion, the Inverse Gaussian describes the time  
required for that motion to hit a fixed distance threshold.
