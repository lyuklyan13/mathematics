## **Theoretical Foundation:**
The von Mises distribution is essentially a continuous "circular analog" of the Normal distribution,  
defined over the cyclical interval $[-\pi, \pi]$. It measures angular or directional data

$$f(x) = \frac{e^{\kappa \cos(x-\mu)}}{2\pi I_0(\kappa)}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import vonmises

kappa_vm = 2.0
dist_vm = vonmises(kappa=kappa_vm)

x_vals = np.linspace(-np.pi, np.pi, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_vm.pdf(x_vals), color='blue', lw=3)
ax1.set_title('Von Mises Circular PDF')
ax2.plot(x_vals, dist_vm.cdf(x_vals), color='blue', lw=3)
ax2.set_title('Von Mises Circular PDF')
plt.show()
```

## **Practical Application:**
Used rigorously in geology to map paleocurrent directions in sedimentary rocks,  
in biology to analyze the homing orientation of migratory sea turtles, and heavily  
in Quantum Information theory to conduct Quantum Tomography on Bloch spheres.

## **Interesting Fact:**
The denominator contains $I_0(\kappa)$, the modified Bessel function of order 0.  
As the concentration parameter $\kappa \to 0$, the distribution flattens into a perfectly uniform   
circular distribution. As $\kappa \to \infty$, it tightens and becomes mathematically indistinguishable  
from a standard Normal distribution wrapping the circle
