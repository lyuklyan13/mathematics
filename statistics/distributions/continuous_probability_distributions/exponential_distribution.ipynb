## **Theoretical Foundation**:
The Exponential distribution models the time elapsed between continuous,  
independent events that occur at a constant average rate $\lambda$.  
$$f(x) = \lambda e^{-\lambda x} \quad \text{for } x \ge 0$$

The mean is exactly $1/\lambda$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

lambda_exp = 1.5
dist_exp = expon(scale=1/lambda_exp)

x_vals = np.linspace(0, 4, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_exp.pdf(x_vals), color='brown', lw=3)
ax1.set_title('Exponential PDF')
ax2.plot(x_vals, dist_exp.cdf(x_vals), color='brown', lw=3)
ax1.set_title('Exponential CDF')
plt.show()
```

## **Practical Application:**
It is ubiquitous in survival analysis, reliability engineering, and risk modeling.  
It successfully predicts the "time to failure" for electronic components that do not  
experience mechanical wear, and the wait times between requests on a cloud server.

## **Interesting Fact:**
The Exponential distribution is the continuous analog of the Geometric distribution    
and is the only continuous distribution exhibiting the "memoryless" property.   
The probability of a machine running for another 10 hours is identical whether  
it was just turned on or has been running for a decade.
