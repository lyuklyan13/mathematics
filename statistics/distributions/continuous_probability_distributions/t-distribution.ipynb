## **Theoretical Foundation:**
The Student's t-distribution arises when estimating the mean of a normally distributed  
population in scenarios where the sample size is small and the population variance is  
fundamentally unknown. It features heavier tails than the normal distribution, controlled  
by the degrees of freedom $\nu$.

$$f(t) = \frac{\Gamma((\nu+1)/2)}{\sqrt{\nu\pi}\,\Gamma(\nu/2)} \left(1+\frac{t^2}{\nu}\right)^{-(\nu+1)/2}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

df_t = 5
dist_t = t(df=df_t)

x_vals = np.linspace(-5, 5, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_t.pdf(x_vals), color='coral', lw=3)
ax1.set_title(f"Student's t-Distribution PDF (df={df_t})")
ax2.plot(x_vals, dist_t.cdf(x_vals), color='coral', lw=3)
ax2.set_title(f"Student's t-Distribution CDF (df={df_t})")
plt.show()
```

## **Practical Application:**
The t-distribution is the absolute standard for A/B testing in tech, pharmacological clinical trials,  
and anywhere experimental data is scarce (e.g., $n < 30$). It provides wider, more conservative  
confidence intervals to account for the uncertainty of estimating variance from a small sample.

## **Interesting Fact:**
The t-distribution represents an elegant mathematical bridge between two extremes:   
when $\nu = 1$, it becomes the pathological Cauchy distribution. As $\nu \to \infty$,  
it morphs seamlessly into the standard Normal distribution.
