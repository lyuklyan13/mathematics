## **Theoretical Foundation:**
In Extreme Value Theory (EVT), while Gumbel Type 1 models right-skewed maxima for standard  
exponential tails, the Gumbel Type 2 (mathematically linked to the Fréchet or Inverse Weibull distribution)  
models extreme maxima specifically drawn from severe heavy-tailed data distributions (like Pareto).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import invweibull

c_frechet = 2.0
dist_frechet = invweibull(c=c_frechet)

x_vals = np.linspace(0.1, 5, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_frechet.pdf(x_vals), color='darkred', lw=3)
ax1.set_title('Gumbel Type 2 (Fréchet) PDF')
ax2.plot(x_vals, dist_frechet.cdf(x_vals), color='darkred', lw=3)
ax2.set_title('Gumbel Type 2 (Fréchet) PDF')
plt.show()
```

## **Practical Application:**
It is heavily deployed in structural engineering, climatology, and hydrology to calculate  
catastrophic risk matrices—specifically predicting the severity of 100-year floods,  
maximum annual precipitation, and peak ocean wave heights that threaten offshore infrastructure.

## **Interesting Fact:**
The Fisher-Tippett-Gnedenko theorem proves that the absolute maximum of a sample    
of identically distributed variables will converge to only one of three possible   
distributions: Gumbel, Fréchet (Gumbel Type 2), or Weibull. Thus, all extreme events  
in the universe ultimately fall into these three mathematical attractors.
