## **Theoretical Foundation:**
The Chi-squared distribution with $k$ degrees of freedom is the continuous  
distribution of the sum of the squares of $k$ independent standard normal random variables.

$$f(x) = \frac{x^{(k/2)-1} e^{-x/2}}{2^{k/2} \Gamma(k/2)} \quad \text{for } x > 0$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

df_chi2 = 4
dist_chi2 = chi2(df=df_chi2)

x_vals = np.linspace(0, 15, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_chi2.pdf(x_vals), color='black', lw=3)
ax1.set_title(f'Chi-squared PDF (df={df_chi2})')
ax2.plot(x_vals, dist_chi2.cdf(x_vals), color='black', lw=3)
ax2.set_title(f'Chi-squared CDF (df={df_chi2})')
plt.show()
```

## **Practical Application:**
It forms the inferential backbone of categorical data science. It is used in the Pearson's  
Chi-squared test for independence in massive contingency tables, assessing goodness-of-fit  
for logistic regression models, and generating confidence bounds for population variance.

## **Interesting Fact:**
The Chi-squared distribution is highly right-skewed at low degrees of freedom,  
but as $k \to \infty$, the skewness evaporates, and the distribution asymptotically  
converges into a Normal distribution, enabling fast numerical approximations for immense datasets.
