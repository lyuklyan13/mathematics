## **Theoretical Foundation:**
A continuous probability distribution with a lower limit $a$, an upper limit $b$,   
and a mode $c$, mapping out a literal triangular shape on a probability density plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import triang

c_tri = 0.3 # Mode
dist_tri = triang(c=c_tri)

x_vals = np.linspace(0, 1, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_tri.pdf(x_vals), color='darkgreen', lw=3)
ax1.fill_between(x_vals, dist_tri.pdf(x_vals), alpha=0.3, color='green')
ax2.set_title('Triangular PDF')
ax2.plot(x_vals, dist_tri.cdf(x_vals), color='darkgreen', lw=3)
ax2.set_title('Triangular CDF')
plt.show()
```

## **Practical Application:**
It is predominantly used in Monte Carlo project management simulations (e.g., PERT estimates)  
where empirical data is lacking, but domain experts can confidently estimate the minimum, maximum,   
and most likely durations for a business task.

## **Interesting Fact:**
When you roll two six-sided dice and sum the values, the discrete probability mass   
shape maps perfectly to a triangle. The continuous analog of that resulting distribution   
is exactly the Triangular distribution.
