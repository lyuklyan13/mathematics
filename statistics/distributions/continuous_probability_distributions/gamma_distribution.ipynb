## **Theoretical Foundation:**
The Gamma distribution is a versatile two-parameter family characterized by a  
shape parameter $k$ (or $\alpha$) and a scale parameter $\theta$. It represents   
the continuous waiting time until the $k$-th Poisson event occurs. 1 

$$f(x) = \frac{1}{\Gamma(k)\theta^k} x^{k-1} e^{-x/\theta}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

alpha_g = 3.0 # Shape
dist_gamma = gamma(a=alpha_g, scale=2.0)

x_vals = np.linspace(0, 20, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_gamma.pdf(x_vals), color='purple', lw=3)
ax1.set_title('Gamma Distribution PDF')
ax2.plot(x_vals, dist_gamma.cdf(x_vals), color='purple', lw=3)
ax1.set_title('Gamma Distribution CDF')
plt.show()
```

## **Practical Application:**
In actuarial science and insurance analytics, it models the aggregate severity  
of financial claim sizes. In climatology, it models highly asymmetric, non-negative  
variables like cumulative precipitation over a season

## **Interesting Fact:**
The Gamma distribution operates as the mathematical master template for multiple distributions:    
an Exponential distribution is just a Gamma with $k=1$, and a Chi-squared distribution  
is a Gamma with $k = \nu/2$ and $\theta = 2$.
