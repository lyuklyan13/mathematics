## **Theoretical Foundation:**
A continuous random variable $X$ follows a lognormal distribution if   
its natural logarithm $Y = \ln(X)$ is normally distributed. It is heavily  
right-skewed and its support is strictly $x \in (0, \infty)$.

$$f(x) = \frac{1}{x \sigma \sqrt{2\pi}} \exp\left(-\frac{(\ln x - \mu)^2}{2\sigma^2}\right)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# s parameter is the standard deviation of the underlying normal distribution
s_log = 0.9 
dist_lognorm = lognorm(s=s_log, scale=np.exp(0))

x_vals = np.linspace(0.01, 5, 1000)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_lognorm.pdf(x_vals), color='teal', lw=3)
ax1.set_title('Lognormal PDF')
ax2.plot(x_vals, dist_lognorm.cdf(x_vals), color='teal', lw=3)
ax2.set_title('Lognormal CDF')
plt.show()
```

## **Practical Application:**
While the normal distribution captures additive effects, the lognormal distribution   flawlessly captures multiplicative growth. It is fundamentally used in quantitative   finance to model asset price paths (Black-Scholes model) and in biology to model cell   growth matrices.

## **Interesting Fact:**
Because it is bounded at zero, the lognormal distribution elegantly avoids the     "negative price" paradox that would occur if one falsely attempted to model stock   prices or property values using a standard Normal distribution.
