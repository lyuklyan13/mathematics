## **Theoretical Foundation:** 
The Weibull distribution acts as a highly flexible generalization of the  
Exponential distribution. By introducing a shape parameter $k$ (or $c$),   
it abandons the memoryless property to model failure rates that change over time

$$f(x) = \frac{k}{\lambda} \left(\frac{x}{\lambda}\right)^{k-1} e^{-(x/\lambda)^k}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

shape_weibull = 1.5 # c parameter
scale_weibull = 1.0 # lambda parameter
dist_weibull = weibull_min(c=shape_weibull, scale=scale_weibull)

x_vals = np.linspace(0, 3, 1000)
fig, ( ax1, ax2 )  = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_weibull.pdf(x_vals), color='darkgreen', lw=3)
ax1.set_title(f'Weibull PDF (k={shape_weibull})')
ax2.plot(x_vals, dist_weibull.cdf(x_vals), color='darkgreen', lw=3)
ax2.set_title(f'Weibull CDF (k={shape_weibull})')
plt.show()
```

## **Practical Application:**
It dominates civil and mechanical engineering. It explicitly models the "bathtub curve"  
of reliability: early "infant mortality" failure, useful life, and end-of-life mechanical wear-out.  
It is also the global standard for predicting wind speed distributions for offshore turbine optimization.

## **Interesting Fact:**
The shape parameter $k$ determines the life-cycle phase. If $k < 1$, the failure rate  
decreases over time (defective manufacturing). If $k = 1$, it reverts exactly to the  
Exponential distribution (random failure). If $k > 1$, the failure rate increases over time (fatigue/wear)
