## **Theoretical Foundation:**
Karl Pearson developed a vast systemic classification of probability distributions via  
differential equations. The Type IV distribution is distinct because it is heavily skewed,   
possesses unbounded support $(-\infty, \infty)$, but uniquely does not possess a closed-form   
cumulative distribution function

$$f(x) \propto \left[1 + \left(\frac{x-\lambda}{a}\right)^2\right]^{-m} \exp\left[-\nu \arctan\left(\frac{x-\lambda}{a}\right)\right]$$

```python
# Pearson Type IV is highly complex and not built natively in scipy.
# We map its unnormalized density kernel.
import numpy as np
import matplotlib.pyplot as plt
def pearson4_kernel(x, m, nu, a, lam):
    return (1 + ((x - lam)/a)**2)**(-m) * np.exp(-nu * np.arctan((x - lam)/a))

x_vals = np.linspace(-5, 5, 1000)
pdf_p4 = pearson4_kernel(x_vals, m=2.5, nu=1.5, a=1, lam=0)

plt.figure(figsize=(6, 5))
plt.plot(x_vals, pdf_p4, color='indigo', lw=3)
plt.title('Pearson Type IV (Unnormalized Kernel)')
plt.show()
```

## **Practical Application:**
Because it allows the decoupled modeling of both extreme skewness and kurtosis jointly,  
it is frequently utilized in advanced operational research to model operating room surgical times,  
asymmetric survival data, and financial return variance predictions.

## **Interesting Fact:**
The required normalization constant for the Pearson Type IV PDF is notoriously difficult  
to calculate because it analytically necessitates the computation of the Complex Gamma Function,  
representing an incredibly rare and elegant intersection between pure complex analysis  
and applied statistics
