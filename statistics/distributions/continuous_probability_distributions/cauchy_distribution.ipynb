## **Theoretical Foundation:**
The Cauchy (or Lorentz) distribution is perfectly symmetric like the Normal  
distribution but possesses incredibly heavy "fat tails". Extreme outliers occur  
with mathematically severe frequency

$$f(x; x_0, \gamma) = \frac{1}{\pi \gamma \left[1 + \left(\frac{x - x_0}{\gamma}\right)^2\right]}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import cauchy

dist_cauchy = cauchy(loc=0, scale=1)

x_vals = np.linspace(-10, 10, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_cauchy.pdf(x_vals), color='magenta', lw=3)
ax1.set_title('Cauchy Distribution PDF')
ax2.plot(x_vals, dist_cauchy.cdf(x_vals), color='magenta', lw=3)
ax2.set_title('Cauchy Distribution PDF')
plt.show()
```

## **Practical Application:**
In machine learning, Cauchy-distributed noise is purposefully injected into stochastic  
optimization algorithms (like simulated annealing) to force the algorithm to randomly  
"jump" out of local minima traps due to its propensity to generate massive outliers.

## **Interesting Fact:**
The Cauchy distribution is pathological in classical statistics: its integral fails  
to converge, meaning it has no defined mean and no defined variance. Consequently,  
calculating the sample average of a million Cauchy variables yields a result just  
as chaotic and dispersed as a single observation, entirely defeating the Law of Large Numbers.

```python
from scipy.stats import norm

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

x = np.linspace(-10, 10, 1000)

axes[0].plot(x, norm.pdf(x), label='Normal distribution (0, 1)', color='blue', linewidth=2)
axes[0].plot(x, cauchy.pdf(x), label='Cauchy distribution (0, 1)', color='red', linewidth=2)
axes[0].set_title('Probability Density Function: Comparing Tails')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Density')
axes[0].legend()
axes[0].grid(True, alpha=0.3)


n_samples = 1000
np.random.seed(13) 

norm_samples = np.random.normal(0, 1, n_samples)
cauchy_samples = np.random.standard_cauchy(n_samples)

axes[1].plot(cauchy_samples, label='Cauchy Sample', alpha=0.7, color='red')
axes[1].plot(norm_samples, label='Normal Sample', alpha=0.9, color='blue')

axes[1].set_title('1000 Random Values (Logarithmic Y-scale)')
axes[1].set_xlabel('Order of Generated Number')
axes[1].set_ylabel('Generated Value')
# axes[1].set_yscale('symlog') 
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
