## **Theoretical Foundation:**
The Normal distribution is the apex of modern statistics, characterized by its perfectly  
symmetrical bell shape. It is parametrized by its mean $\mu$ (center) and variance $\sigma^2$ (dispersion).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu_norm, sigma_norm = 0, 1
dist_norm = norm(loc=mu_norm, scale=sigma_norm)

print(f'Skewness:\t {dist_norm.stats(moments="s")} | Kurtosis:\t {dist_norm.stats(moments="k")}')

x_vals = np.linspace(-4, 4, 1000)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_norm.pdf(x_vals), 'k-', lw=3)
ax1.fill_between(x_vals, dist_norm.pdf(x_vals), alpha=0.2, color='gray')
ax1.set_title('Standard Normal PDF')

ax2.plot(x_vals, dist_norm.cdf(x_vals), 'k-', lw=3)
ax2.set_title('Standard Normal CDF')
plt.show()
```

## **Practical Application:**
Due to the Central Limit Theorem (CLT), it models the aggregation of independent random errors.  
It is universally applied in machine learning for Ordinary Least Squares (OLS) regression assumptions,  
Kalman filtering in robotics, and deep learning weight initializations.

## **Interesting Fact:**
The Normal distribution is the mathematical embodiment of maximum entropy on the entire real  
line $\mathbb{R}$ when only the mean and variance are known. Furthermore, it is the only distribution  
whose cumulants beyond variance are exactly zero.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set ggplot style for an R-like appearance
plt.style.use('ggplot')

# Define parameters for the Standard Normal Distribution
mu = 0
sigma = 1

# Generate x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Create the figure
fig, ax = plt.subplots(figsize=(12, 7))

# Plot the main curve
ax.plot(x, y, color='black', linewidth=2)


# 1 Standard Deviation (68%)
x1 = np.linspace(mu - sigma, mu + sigma, 100)
ax.fill_between(x1, norm.pdf(x1, mu, sigma), color='#348ABD', alpha=0.8)

# 2 Standard Deviations (95%)
x2_left = np.linspace(mu - 2*sigma, mu - sigma, 100)
x2_right = np.linspace(mu + sigma, mu + 2*sigma, 100)
ax.fill_between(x2_left, norm.pdf(x2_left, mu, sigma), color='#E24A33', alpha=0.8)
ax.fill_between(x2_right, norm.pdf(x2_right, mu, sigma), color='#E24A33', alpha=0.8)

# 3 Standard Deviations (99.7%)
x3_left = np.linspace(mu - 3*sigma, mu - 2*sigma, 100)
x3_right = np.linspace(mu + 2*sigma, mu + 3*sigma, 100)
ax.fill_between(x3_left, norm.pdf(x3_left, mu, sigma), color='#988ED5', alpha=0.8)
ax.fill_between(x3_right, norm.pdf(x3_right, mu, sigma), color='#988ED5', alpha=0.8)

# ADD TEXT INSIDE THE CURVE
ax.text(-0.5, 0.15, '34.1%', ha='center', color='white', fontweight='bold', fontsize=12)
ax.text(0.5, 0.15, '34.1%', ha='center', color='white', fontweight='bold', fontsize=12)

ax.text(-1.5, 0.05, '13.6%', ha='center', color='white', fontweight='bold', fontsize=11)
ax.text(1.5, 0.05, '13.6%', ha='center', color='white', fontweight='bold', fontsize=11)

ax.text(-2.5, 0.015, '2.1%', ha='center', color='white', fontweight='bold', fontsize=10)
ax.text(2.5, 0.015, '2.1%', ha='center', color='white', fontweight='bold', fontsize=10)


# 68% Arrow
ax.annotate('', xy=(-1, 0.25), xytext=(1, 0.25), arrowprops=dict(arrowstyle="<->", color='black', lw=1.5))
ax.text(0, 0.26, '68%', ha='center', va='bottom', fontsize=12, fontweight='bold')

# 95% Arrow
ax.annotate('', xy=(-2, 0.08), xytext=(2, 0.08), arrowprops=dict(arrowstyle="<->", color='black', lw=1.5))
ax.text(0, 0.09, '95%', ha='center', va='bottom', fontsize=12, fontweight='bold')

# 99.7% Arrow
ax.annotate('', xy=(-3, 0.02), xytext=(3, 0.02), arrowprops=dict(arrowstyle="<->", color='black', lw=1.5))
ax.text(0, 0.03, '99.7%', ha='center', va='bottom', fontsize=12, fontweight='bold')


ax.set_title('Normal Distribution: The 68-95-99.7 Rule', fontsize=16, fontweight='bold', pad=20)
ax.set_ylabel('Probability Density', fontsize=12)

# Customize X-axis to show standard deviations (mu, sigma)
ax.set_xticks([-3, -2, -1, 0, 1, 2, 3])
ax.set_xticklabels([r'$-3\sigma$', r'$-2\sigma$', r'$-1\sigma$', r'$\mu$', r'$+1\sigma$', r'$+2\sigma$', r'$+3\sigma$'], fontsize=14)

# Remove the default grid for a cleaner look
ax.grid(False)

plt.tight_layout()
plt.show()
```

**The 68–95–99.7 Rule**

These are the magic numbers of statistics. In any normal distribution:  

68% of all values ​​lie within 1 standard deviation of the mean.  
95% of values ​​lie within 2 standard deviations. (This is considered the "normal range").  
99.7% of values ​​lie within 3 standard deviations. Anything outside these limits  
is a great rarity (Outliers).
