## **Theoretical Foundation:**
The Non-central t-distribution shifts the symmetrical t-distribution by introducing   
a noncentrality parameter $\mu$, reflecting the case where the underlying true mean is non-zero.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nct

df_nct, nc_nct = 10, 2.0
dist_nct = nct(df=df_nct, nc=nc_nct)

x_vals = np.linspace(-2, 7, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_nct.pdf(x_vals), color='salmon', lw=3)
ax1.set_title(f'Non-central t PDF (nc={nc_nct})')
ax2.plot(x_vals, dist_nct.cdf(x_vals), color='salmon', lw=3)
ax1.set_title(f'Non-central t CDF (nc={nc_nct})')
plt.show()
```

## **Practical Application:**
While central t-distributions evaluate the null hypothesis, the Non-central t is  
strictly utilized to calculate statistical power (the likelihood of correctly rejecting  
a false null hypothesis) and to establish asymmetrical confidence intervals around effect  
sizes in robust experimental design.

## **Interesting Fact:**
Unlike the standard t-distribution, the Non-central t is intrinsically skewed.   
    The center of mass and the direction of the skew interact in complex ways   
depending on the precise ratio between the degrees of freedom and the noncentrality parameter.
