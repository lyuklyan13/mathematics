## **Theoretical Foundation:**
The F-distribution characterizes the continuous ratio of two scaled, independent  
Chi-squared random variables. It relies on two distinct degrees of freedom:  
$d_1$ (numerator) and $d_2$ (denominator)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

dfn, dfd = 5, 10
dist_f = f(dfn=dfn, dfd=dfd)

x_vals = np.linspace(0, 5, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_f.pdf(x_vals), color='sienna', lw=3)
ax1.set_title(f'Fisher F-Distribution PDF')
ax2.plot(x_vals, dist_f.cdf(x_vals), color='sienna', lw=3)
ax2.set_title(f'Fisher F-Distribution CDF')
plt.show()
```

## **Practical Application:**
It is the probabilistic core of Analysis of Variance (ANOVA). Data scientists use  
it to test if the means of multiple distinct clusters are statistically equivalent  
by comparing the ratio of between-group variance to within-group variance.

## **Interesting Fact:**
The F-distribution has an inherent reciprocal symmetry: if $X \sim F(d_1, d_2)$,  
then its mathematical reciprocal $1/X$ seamlessly follows $F(d_2, d_1)$, a property   
vastly leveraged in complex variance component testing.
