## **Theoretical Foundation:**
The Pareto distribution describes a power-law probability where the tail decreases  
linearly on a logarithmic scale. It depends on a shape parameter $\alpha$  
and a minimum threshold $x_m$.  

$$f(x) = \frac{\alpha x_m^\alpha}{x^{\alpha+1}} \quad \text{for } x \ge x_m$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

b_pareto = 2.62
dist_pareto = pareto(b=b_pareto)

x_vals = np.linspace(1, 5, 1000)
fig, ( ax1, ax2 ) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_pareto.pdf(x_vals), color='gold', lw=3)
ax1.set_title('Pareto Distribution PDF')
ax2.plot(x_vals, dist_pareto.cdf(x_vals), color='gold', lw=3)
ax1.set_title('Pareto Distribution CDF')
plt.show()
```

## **Practical Application:**
It mathematically codifies the Pareto Principle (the 80/20 rule). It is deployed  
in economics to model extreme wealth inequality, where a tiny fraction of the  
population commands the vast majority of capital, and in network science to model  
the extreme scale-free connections of the internet.

## **Interesting Fact:**
Depending on the specific scale of the shape parameter $\alpha$, the Pareto distribution   
can have an infinite variance ($\alpha \le 2$) or even an infinite mean ($\alpha \le 1$),  
perfectly demonstrating why average-based metrics (like GDP per capita) fail catastrophically  
when analyzing highly skewed socioeconomic inequality.
