## **Theoretical Foundation:**
The Rice distribution is a generalization of the Rayleigh distribution,  
modeling the magnitude of a circular bivariate normal random variable with  
a non-zero mean. It features a noncentrality parameter.

$$f(x) = \frac{x}{\sigma^2} \exp\left(-\frac{x^2+\nu^2}{2\sigma^2}\right) I_0\left(\frac{x\nu}{\sigma^2}\right)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rice

b_rice = 2.0
dist_rice = rice(b=b_rice)

x_vals = np.linspace(0, 8, 1000)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(x_vals, dist_rice.pdf(x_vals), color='darkcyan', lw=3)
ax1.set_title(f'Rice Distribution PDF (b={b_rice})')

ax2.plot(x_vals, dist_rice.cdf(x_vals), color='darkcyan', lw=3)
ax2.set_title(f'Rice Distribution CDF (b={b_rice})')

plt.show()
```

## **Practical Application:**
It is ubiquitous in wireless communications theory (4G, 5G). While Rayleigh fading models  
communication channels with multiple scattered signal paths but no direct Line-of-Sight (LOS),  
Rician fading models channels where a strong, direct LOS signal dominates alongside scattered  
multipath echoes

## **Interesting Fact:**
When the Line-of-Sight signal strength drops to absolute zero ($\nu \to 0$),  
the complex Bessel function in the Rice distribution collapses, and the distribution  
mathematically reduces exactly back to the Rayleigh distribution.
