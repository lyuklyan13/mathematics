## **Theoretical Foundation:**
A generalization of the Chi distribution, the Non-central Chi models the Euclidean  
norm of $k$ independent normal variables that have a non-zero mean. It is defined  
by degrees of freedom $k$ and a noncentrality parameter $\lambda = \sqrt{\mu_1^2 + \dots + \mu_k^2}$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ncx2

df_ncchi = 3
nc_param = 2.0
# Derived mathematically from the non-central chi-squared (ncx2)
x_vals = np.linspace(0.01, 8, 1000)
pdf_ncchi = 2 * x_vals * ncx2.pdf(x_vals**2, df=df_ncchi, nc=nc_param**2)

plt.figure(figsize=(6, 5))
plt.plot(x_vals, pdf_ncchi, color='steelblue', lw=3)
plt.title('Non-central Chi PDF')
plt.show()
```

## **Practical Application:**
It is paramount in telecommunications and signal processing. It models the amplitude  
envelope of an RF signal that contains both a strong, deterministic, shifted component  
(the noncentrality) and superimposed Gaussian background noise.

## **Interesting Fact:**
When the degrees of freedom $k = 2$, the Non-central Chi distribution maps perfectly  
and analytically to the Rice (Rician) fading distribution used in mobile networking.
