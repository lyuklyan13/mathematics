Рівняння прямої  
y' = w * x +  b;  
y' - наше передбачення  
w - нахил прямої  
b - зсув по осі OY  


Згідно з методом найменших квадратів **наша мета**  
це — знайти такі коефіцієнти $\alpha_0$ та $\alpha_1$, при яких  
функція  $S$ (середнє квадратичне відхилення) буде мінімальною:   
$$S(\alpha_0, \alpha_1) = M(Y - \alpha_1 X - \alpha_0)^2 \rightarrow \min$$

Щоб знайти мінімум функції двох змінних, потрібно знайти її  
частинні похідні по кожній з цих змінних ($\alpha_0$ та $\alpha_1$)  
і прирівняти   їх до нуля.

Розв'язавши відповідну ( не складу ) математичну задачу  
отримуємо:

$$\alpha_0 = M(Y) - \alpha_1 M(X)$$

$$\alpha_1 = \frac{cov(X,Y)}{\sigma_x^2}$$

У нашому випадку:  
$$\alpha_1 = w $$

У більш зручному вигляді для обрахунків:

$$w = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}$$

$$b = \bar{y} - w \cdot \bar{x}$$

Тоді можна реагізовувати на Python

```python
import numpy as np
import matplotlib.pyplot as plt 


X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 3.5, 4, 4.5, 7])


def fit_linear_regression(X, Y):
    """Функція для знаходження коефіцієнтів лінійної регресії (w, b)"""

    x_mean = np.mean(X)
    y_mean = np.mean(Y)

    numerator = np.sum((X - x_mean)* (Y - y_mean))
    denominator = np.sum((X - x_mean) ** 2)

    w = numerator / denominator
    b = y_mean - w * x_mean

    return w, b

def predict(X, w, b):
    return w * X + b


w, b  = fit_linear_regression(X, Y)

Y_pred = predict(X, w, b)

# Візуалізація результатів

plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Дані')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Наша лінія регресії')


plt.xlabel('X')
plt.ylabel('Y')
plt.title('Проста лінійна регресія з нуля')
plt.legend()
plt.title('Проста лінійна регресія з нуля')
plt.legend()
plt.grid()
plt.show()


```

```python
new_x = 6
new_y = predict(new_x, w, b)
print(f"Прогноз для X={new_x}: Y={new_y}")
```

```python
import numpy as np
X = np.array(np.random.rand(100) * 10)
Y = 2.5 * X + 1.0 + np.random.normal(0, 2, size=X.shape)

w, b  = fit_linear_regression(X, Y)

Y_pred = predict(X, w, b)

# Візуалізація результатів

plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Дані')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Наша лінія регресії')


plt.xlabel('X')
plt.ylabel('Y')
plt.title('Проста лінійна регресія з нуля')
plt.legend()
plt.title('Проста лінійна регресія з нуля')
plt.legend()
plt.grid()
plt.show()
```
